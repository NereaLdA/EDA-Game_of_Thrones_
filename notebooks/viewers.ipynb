```python
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_viewers = pd.read_csv('../data/viewers.csv')
df_viewers
```

```python
df_viewers.info()
```

```python
df_viewers_clean = df_viewers.drop(["Original air date [20]"], axis=1)
df_viewers_clean
```

```python
df_viewers_clean.to_csv("../data_clean/df_viewers_clean.csv", index=False)
```

```python
plt.plot(df_viewers_clean['No.overall'], df_viewers_clean['U.S. viewers(millions)'])
plt.xlabel("Episodio (global)")
plt.ylabel("Audiencia (millones)")
plt.title("Audiencia por episodio")
plt.show()
```

audiencia media por temporada:

```python
df_viewers_clean.groupby('Season')['U.S. viewers(millions)'].mean().plot(kind='bar', title='Audiencia media por temporada')
```

```python
top10 = df_viewers_clean.sort_values(by='U.S. viewers(millions)', ascending=False).head(10)
top10[['Title', 'U.S. viewers(millions)']].plot(kind='bar', x='Title', title='Top 10 episodios más vistos')
```

```python
top10 = df_viewers_clean.sort_values(by='U.S. viewers(millions)', ascending=True).head(10)
top10[['Title', 'U.S. viewers(millions)']].plot(kind='bar', x='Title', title='Top 10 episodios menos vistos')
```
