```python
#Mergear muertes x episodio (por episode) usar  method_cat 2 y episodes (por no of episode overall) usar imdb rating

df_muertes_epi2_merge = pd.read_csv('../data/muertes_x_episodio2.csv')
df_muertes_epi2_merge
```

```python
df_muertes_epi2_merge_clean = df_muertes_epi2_merge.drop(["order","season","character_killed","killer","method","reason","location","importance","allegiance"], axis=1)
df_muertes_epi2_merge_clean
```

```python
df_viewers_merge = pd.read_csv('../data/viewers.csv')
df_viewers_merge
```

```python
df_viewers_merge_clean = df_viewers_merge.drop(["Unnamed: 0", "No.overall", "Title", "Directed by", "Written by", "Original air date [20]", "Season"],axis=1)
df_viewers_merge_clean
```

```python
df_viewers_merge_clean.rename(columns={"No. inseason": "episode"}, inplace=True)
df_viewers_merge_clean
```

```python
# Join many-to-one: se añaden los datos de viewers al dataframe de muertes por episodio
df_muertes_y_viewers = df_muertes_epi2_merge_clean.merge(df_viewers_merge_clean, on='episode', how='left')

df_muertes_y_viewers
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar número de muertes por episodio y método
df_grouped = df_muertes_y_viewers.groupby(['episode', 'method_cat']).size().reset_index(name='muertes_por_metodo')

# Añadir la audiencia de cada episodio (una única vez por episodio)
df_grouped = df_grouped.merge(
    df_muertes_y_viewers[['episode', 'U.S. viewers(millions)']].drop_duplicates(),
    on='episode',
    how='left'
)

# Crear gráfico
plt.figure(figsize=(14, 8))

sns.scatterplot(
    data=df_grouped,
    x='episode',
    y='muertes_por_metodo',
    hue='method_cat',
    size='U.S. viewers(millions)',
    sizes=(20, 400),
    alpha=0.7,
    palette='tab10'
)

plt.title('Métodos de muerte por episodio y audiencia')
plt.xlabel('Episodio')
plt.ylabel('Número de muertes por método')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Asegúrate de tener este DataFrame:
# df con columnas: 'episode', 'method_cat', 'U.S. viewers(millions)'

# Crear paleta GOT ampliada
GOTcolors = [
    "#1C1C1C", "#8B0000", "#4682B4", "#DAA520", "#708090",
    "#013220", "#E0FFFF", "#FFFAF0", "#2F4F4F", "#A52A2A",
    "#4B0082", "#C0C0C0", "#5D3A00", "#191970", "#800000", "#556B2F"
]

# Pivot para la barra apilada
pivot_df = df_muertes_y_viewers.groupby(['episode', 'method_cat']).size().unstack(fill_value=0)
viewers_by_episode = df_muertes_y_viewers.groupby('episode')["U.S. viewers(millions)"].mean()

# Asegurar colores suficientes
num_methods = len(pivot_df.columns)
colors = GOTcolors[:num_methods] if num_methods <= len(GOTcolors) else GOTcolors + ['#777777'] * (num_methods - len(GOTcolors))

# Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(16, 8))

# Gráfico de barras apiladas
pivot_df.plot(kind='bar', stacked=True, ax=ax1, color=colors)
ax1.set_xlabel("Episodio")
ax1.set_ylabel("Número de muertes")
ax1.set_title("MÉTODOS MUERTE POR EPISODIO Y AUDIENCIA")
ax1.tick_params(axis='x', rotation=45)

# Eje secundario para viewers
ax2 = ax1.twinx()
ax2.plot(viewers_by_episode.index, viewers_by_episode.values, color='gold', marker='o', linewidth=2, label="Viewers (millones)")
ax2.set_ylabel("Viewers (millones)")
ax2.legend(loc='upper right')

# Leyenda de los métodos
ax1.legend(title="Método de muerte", bbox_to_anchor=(1.15, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Paso 1: obtener los 5 episodios con más audiencia
top_5_episodios = df_muertes_y_viewers.groupby("episode")["U.S. viewers(millions)"].mean().nlargest(5).index

# Paso 2: filtrar el DataFrame original
top_df = df_muertes_y_viewers[df_muertes_y_viewers["episode"].isin(top_5_episodios)]

# Paso 3: contar los métodos de muerte en esos episodios
metodos_top5 = top_df["method_cat"].value_counts().head(5)

# Paso 4: graficar
plt.figure(figsize=(8,5))
metodos_top5.plot(kind="bar", color=GOTcolors)
plt.title("Top 5 métodos más usados en los 5 episodios más vistos")
plt.xlabel("Método de muerte")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
